```python
#function 39
df1 = pd.DataFrame({'Pratik_left': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'Pratik_right': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
df1.merge(df2, left_on='Pratik_left', right_on='Pratik_right')
```

```python
#function 40
df1.join(df2 , lsuffix = "_caller" , rsuffix = "_other")
```
